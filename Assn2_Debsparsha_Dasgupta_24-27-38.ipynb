Question 3

```python
import numpy as np
import numpy.linalg as la

#a
def vandermonde(N):
    vec = np.arange(N) + 1
    vander = vec ** np.arange(N).reshape(-1, 1)
    return vander.T

N = 12
vander = vandermonde(N)
print("Vandermonde Matrix for N=12:\n", vander)

#b
x = np.ones(N)
b = np.dot(vander, x)
print("\nVector b:\n", b)

#c
V_inv = la.inv(vander)
x_naive = np.dot(V_inv, b)
print("\nSolution using matrix inversion:\n", x_naive)

#d
x_solve = la.solve(vander, b)
print("\nSolution using solve:\n", x_solve)
```

The result from solve is much closer to the actual value of x (all ones), while the result from matrix inversion deviates due to numerical instability.
